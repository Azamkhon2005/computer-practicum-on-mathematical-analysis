```python
x, y, z = symbols("x y z")
def f(x,n):
  res = x
  for i in range(1, n+1):
    res *= (x+i)
  return res
print(f(6, 7))
print(f(x, 4))
try:
  print(f(y+z, x))
except:
  print("Не возможно выполнить операцию для аргументов y+z, x ")
```

#### Задание 3
Опиcать функцию $square\_roots$, возвращающую корни квадратного уравнения $a x^2 + bx + c = 0$  с вещественными аргументами $a\neq 0$, b, c и  необязательным параметром $is\_real$, по умолчанию равным $True$. Функция возвращает список корней уравнения. Если корни комплексные, а $is\_real=True$, то функция возвращает пустой список. Если корень один, то в списке корней он тоже должен быть один.

Решить уравнения в вещественных числах:

а) $3x^2+8x+4=0$,

б) $x^2-4x+4=0$

в) $2x^2+x+1=0$

Уравнение в) решить отдельно в комплексных числах.

Для красивого представления комплексных чисел можно воспользоваться функцией display.

Если элементы списка нужно выводить в столбик, можно воспользоваться display(\*имя_списка) (\* в данном случае служебный символ)

```python
def square_roots(a, b, c, is_real=True):
    # Вычисляем дискриминант
    discriminant = b**2 - 4*a*c

    # Проверяем, являются ли корни вещественными
    if discriminant > 0 and is_real:
        # Вычисляем корни
        root1 = (-b + sqrt(discriminant)) / (2*a)
        root2 = (-b - sqrt(discriminant)) / (2*a)

        # Возвращаем корни в виде списка
        return [float(root1), float(root2)]

    elif discriminant==0:
      return [float(-b / 2*a)]

    elif discriminant<0 and not is_real:
      root1 = (-b + I*sqrt(abs(discriminant))) / (2*a)
      root2 = (-b - I*sqrt(abs(discriminant))) / (2*a)
      # Возвращаем корни в виде списка
      return [root1, root2]
    else:
        # Возвращаем пустой список, так как корни комплексные, а is_real=True
        return []
# a)  3x^2+8x+4=0
x1, x2 = square_roots(3,8,4)
print("корни 3x^2+8x+4=0:")
print(f"x1 = {x1}, x2 = {x2}")
print('-'*30)
# б)  x2−4x+4=0
print("корни x^2−4x+4=0")
x1 = square_roots(1,-4,4)[0]
print(f"x1 = {x1}")
print('-'*30)
# c)  2x^2+x+1=0

x1, x2 = square_roots(2,1,1,False)
print("корни 2x^2+1x+1=0:")
display(x1,x2)
```

#### Задание 4
Описать функцию $less\_c(my\_list,c)$, возвращающую первый попавшийся элемент списка $my\_list$, меньший $c$, и его номер в списке. Если таких элементов нет, возвращается ($None, None$).

Вызвать функцию $less\_c(my\_list,c)$ со списком $my\_list = [3, -2, 8]$ и $c = - 4$, затем с тем же списком, но $c = 1$.

С помощью этой функции найти четное натуральное число $n$ до 100 (включительно), такое что $\sin n <\frac12$, построив предварительно список $list\_sin$ синусов четных натуральных чисел (строить список с помощью for и range).

```python
def less_c(my_list, c):
    for index, element in enumerate(my_list):
        if element < c:
            return element, index
    return None, None

# my_list=[3, -2, 8] и c=-4
result1 = less_c([3, -2, 8], -4)
print("Результат для c=-4:", result1)

# my_list=[3, -2, 8] и c=1
result2 = less_c([3, -2, 8], 1)
print("Результат для c=1:", result2)

list_sin = [np.sin(2*n) for n in range(1, 51)]

# Поиск четного натурального числа n, такого что sin(n) < 0.5
for n, sin_value in enumerate(list_sin):
    if sin_value < 0.5 and n % 2 == 0:
        print(f"Первое четное натуральное число n, для которого sin(n) < 0.5: {n}")
        break
```

#### Задание 5
Построить на отрезке $[-2\pi,2\pi]$ график $tg(x)$ зеленым цветом, а график $ctg(x)$ красным, задать предел по вертикальной оси (-5, 5).

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y_tan = np.tan(x)
y_cot = 1/np.tan(x)
plt.figure(figsize=(8, 6))
plt.plot(x, y_tan, label='tg(x)', color='green')
plt.plot(x, y_cot, label='ctg(x)', color='red')
plt.ylim(-5, 5)
plt.legend()
plt.title('Графики функций tg(x) и ctg(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

#### Задание 7
Изобразить на графике гиперболу, заданную своим каноническим уравнением
$$
\frac{x^2}{a^2}-\frac{y^2}{b^2}=1,
$$
при $a=3$, $b=5$. Масштаб по осям одинаковый, рекомендуемый интервал (-10, 10).

```python
a = 3
b = 5

# Генерация значений x для построения графика
x = np.linspace(-10, 10, 400)
y_pos = np.sqrt((x**2 * a**2 - a**2) / b**2)
y_neg = -y_pos
plt.figure(figsize=(8, 6))
plt.plot(x, y_pos,color="blue")
plt.plot(x, y_neg,color="blue")

plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.legend()
plt.title('Гипербола: $\\frac{x^2}{a^2} - \\frac{y^2}{b^2} = 1$, при $a=3, b=5$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
