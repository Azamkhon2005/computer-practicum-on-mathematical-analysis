### Индивидуальное задание
Построить линии уровня и найти полный дифференциал функции
$f(x, y)$.

Вариант $31$


$f(x, y) = x^{2} + y + \frac{2 \sqrt{3} \sqrt{- x y}}{9}$

Построить линии уровня для значений: $8$ - синим цветом, $3$ - черным цветом, $30$ - красным цветом


```python
C = symbols('C', real=True)
def contourline(f, x, y, C):
 return sympy.Eq(f(x, y), C)
def f(x, y):
  return x**2 + y + (2*(3**(S(1)/2)) * (-x*y)**(S(1)/2))/9
p = plot(ylabel='y', show=False)

for item in [(8, 'blue'),  (3, 'black'), (30, 'red')]:
 c, col = item
 p.append(plot_implicit(contourline(f, x, y, c),(x, -10, 10), (y, -10, 10),  line_color=col, line_width=15, show=False)[0])

p.show()
```
