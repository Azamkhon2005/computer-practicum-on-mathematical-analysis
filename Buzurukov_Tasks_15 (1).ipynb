### Задание 1
Найти точки экстремума:

1) $z = x^2 + xy + y^2 - 2x - y$

2) $z = x^3y^2(6 - x - y)$, $(x, y > 0)$

3) $z = \frac{8}{x} + \frac{x}{y} + y$

4) $z = \frac{1 + x - y}{\sqrt{1 + x^2 + y^2}}$
###### Указание.
Вначале найти стационарные точки. Составить определитель из вторых производных в произвольной точке. Подставляя найденные (вещественные) стационарные точки, решить вопрос о наличии экстремума.

```python
x, y = symbols("x y", positive = True)
def find_extremum(f):

  stat_points = nonlinsolve([f.diff(x), f.diff(y)], [x, y])
  Delta = Matrix([[f.diff(x, 2), f.diff(x, y)], [f.diff(x, y), f.diff(y, 2)]]).det()
  if type(stat_points) == sympy.sets.sets.FiniteSet:
    prep = list(stat_points)
  else:
    return
  for i in prep:
      try:
        if Delta.subs({x: float(i[0]), y: float(i[1])}) > 0:
          print(f"Точки экстремума :", i[0], i[1])
      except:
        pass
arr = [x**2 + x*y + y**2 - 2*x - y, (x**3)*(x**2)*(6 - x - y),8/x + x/y + y ,(1 + x - y)/ (1 + x**2 + y**2)**0.5 ]
for i in range(len(arr)):
  print("Для функции ")
  display(arr[i])
  find_extremum(arr[i])
  print("**"*10)
```

### Задание 2.
С помощью метода Лагранжа найти условный экстремум функции
$z = x + 2y$ при условии $x^2 + y^2 = 5$.

```python
x, y, lam = symbols('x y lamda')# нужны символы без ограничений по знаку!
varrs = [x, y, lam]
z = 6 - 4*x - 3*y
restriction = x**2 + y**2 - 5
L = z + lam*restriction
stat_points = nonlinsolve([L.diff(var)for var in varrs], varrs)
display(stat_points)
```

```python
dx, dy, dx2, dy2 = symbols('dx dy dx^2 dy^2')
d2L = L.diff(x,2)*dx2 + 2*L.diff(x,y)*dx*dy + L.diff(y,2)*dy2
factor(d2L)
```

при $\lambda$ > 0 второй дифференциал положителен, при $\lambda$ < 0 отрицателен, поэтому точка $(-4\sqrt{5}/5,-3\sqrt{5}/5)$ - точка локального
максимума, $(4\sqrt{5}/5,3\sqrt{5}/5)$ - точка локального минимума.
Следовательно, точки $(-4\sqrt{5}/5,-3\sqrt{5}/5)$ и $(4\sqrt{5}/5,3\sqrt{5}/5)$-условные экстремумы функции.

### Задание 4.
С помощью метода Лагранжа найти экстремум функции $u = x - 2y + 2z$ при условии $x^2 + y^2 + z^2 = 9$.

```python
x, y, z, lam = symbols('x y z lamda')
varrs = [x, y, z, lam]
u = x - 2*y + 2*z
restriction = x**2 + y**2 + z**2 - 9
L = u + lam*restriction
stat_points = nonlinsolve([L.diff(var)for var in varrs], varrs)
display(stat_points)
dx, dy, dz, dx2, dy2, dz2 = symbols('dx dy dz dx^2 dy^2 dz^2')
d2L = 0
diffs = {x: (dx, dx2), y: (dy, dy2), z: (dz, dz2)}
for i, var in enumerate(varrs[:-1]):
  d2L += L.diff(var, 2)*diffs[var][1]
  for j in range(i + 1, 3):
    d2L += 2*L.diff(var, varrs[j])*diffs[var][0]*diffs[varrs[j]][0]
display(d2L)
```

при $\lambda$ > 0 второй дифференциал положителен, при $\lambda$ < 0 отрицателен, поэтому точка $(1, -2, 2)$ - точка локального
максимума, $(-1, 2, -2)$ - точка локального минимума.
Следовательно, точки $(1, -2, 2)$ и $(-1, 2, -2)$ - условные экстремумы функции.
