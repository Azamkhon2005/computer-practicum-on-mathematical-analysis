```python
import numpy as np
import sympy
from sympy import nsolve,diff, lambdify
import matplotlib.pyplot as plt
%matplotlib inline
```

### –ó–∞–¥–∞—á–∞ 1
–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ $ùë•^4‚àíe^x=0$ –ø—Ä–∏ $-2 \le x \le 2$, –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫ –ª–µ–≤–æ–π —á–∞—Å—Ç–∏, –æ—Ç–º–µ—Ç–∏—Ç—å —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è —Å –æ—Å—å—é –û–•.

```python
from sys import modules
from sympy.abc import x # –≠—Ç–æ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–π —Å–ø–æ—Å–æ–± –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å–∏–º–≤–æ–ª–∞.
display(nsolve(x**4 - sympy.E**x, x, 2))
display(nsolve(x**4 - sympy.E**x, x, -2))

X = np.linspace(-2, 0, 100)
plt.plot(X, X**4 - sympy.E**X)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
plt.ylim(-1, 1)
```

### –ó–∞–¥–∞—á–∞ 2.
–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ $\sin^2(2x)=\ln x$ –ø—Ä–∏ $0< x \le 3$. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –ª–µ–≤–æ–π –∏ –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏, –æ—Ç–º–µ—Ç–∏—Ç—å —Ç–æ—á–∫—É –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤.

```python
display(nsolve(sympy.sin(2*x)**2 - sympy.ln(x), x, 3))
display(nsolve(sympy.sin(2*x)**2 - sympy.ln(x), x, 0.1))
X = np.linspace(0.1, 3, 100)
expr = diff(sympy.sin(2*x)**2 - sympy.ln(x), x)
f = lambdify(x, expr)
ax = plt.gca()
plt.plot(X,f(X))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
plt.ylim(-3, 3)
```

### –ó–∞–¥–∞—á–∞ 3.
–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ $\sin^2(2x^2-7x+6)=3/4$ –Ω–∞ $[3, 4]$. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –ª–µ–≤–æ–π –∏ –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏, –æ—Ç–º–µ—Ç–∏—Ç—å —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤.

```python
display(nsolve(sympy.sin(2*x**2 - 7*x + 6)**2 , x, 4))
display(sympy.solve(sympy.sin(2*x**2 - 7*x + 6)**2 , x))
X = np.linspace(0, 4, 100)
expr = diff(sympy.sin(2*x**2 - 7*x + 6)**2, x)
f = lambdify(x, expr)
ax = plt.gca()
ax.plot(X,f(X))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
```
